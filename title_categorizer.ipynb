```python
import pandas as pd

# Load the datasets
df1 = pd.read_csv(r'C:\Users\abhinavadhikari\BusinessForecastingProject-main\BusinessForecastingProject-main\amazon_products.csv')
df1.rename(columns={'category_id': 'id'}, inplace=True)
df2 = pd.read_csv(r'C:\Users\abhinavadhikari\BusinessForecastingProject-main\BusinessForecastingProject-main\amazon_categories.csv')

# Merge the datasets
merged_df = pd.merge(df1, df2, on='id')

# Group by category_name and count the number of titles in each category
category_counts = merged_df.groupby('category_name')['title'].count().reset_index()
category_counts.columns = ['category_name', 'title_count']

# Output the category counts for reference
print("Category counts:")
print(category_counts)

# Create a dataframe to store top 20 titles for each category
top_20_titles_per_category = []

# Loop through each category and extract top 20 titles
for category_name in category_counts['category_name']:
    category_titles = merged_df[merged_df['category_name'] == category_name]
    top_titles = category_titles['title'].value_counts().head(20).reset_index()
    top_titles.columns = ['title', 'count']  # Rename columns
    top_titles = top_titles[['title']]  # Retain only the title column
    top_titles['category_name'] = category_name

    # Append to the list
    top_20_titles_per_category.append(top_titles)

# Combine all top 20 titles into a single dataframe
top_20_titles_df = pd.concat(top_20_titles_per_category, ignore_index=True)

# Save the top 20 titles for all categories to a CSV file
top_20_titles_df.to_csv("top_20_titles_per_category.csv", index=False)

print("Top 20 titles for all categories have been saved.")
```
